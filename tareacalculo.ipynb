tarea param√©tricas

```python
from sympy import symbols, sin, cos, pi
from sympy import plot_parametric
t = symbols('t')
def x(t):
  return (cos(t)+(1/2)*(cos(7*t))+(1/3)*sin(17*t))
def y(t):
  return (sin(t)+(1/2)*(sin(7*t))+(1/3)*cos(17*t))

plot_parametric((x(t), y(t), (t, 0, 14*pi)));
```

tarea grafica 1

```python
from sympy import symbols, Eq
from sympy import plot_implicit
x, y = symbols('x y')
plot_implicit(Eq(x**2+y**3-2*y-3, 0), (x, -15, 15),
                 (y, -15, 15));
```

tarea grafica 2

```python
from sympy import symbols, Eq
from sympy import plot_implicit
import numpy as np
paso = .01
y = np.arange(-1, 1, paso)
# quite el cero
y = y[y!=0]
x, y = symbols('x y')
plot_implicit(Eq((x+(5/(y**2)))*(y-(3/100)) -9.7, 0), (x, -15, 15),
                  (y, -15, 15));
```

grafica polares

```python
from re import A
from sympy import symbols
import numpy as np
import matplotlib.pyplot as plt


theta = np.linspace(0, 2*np.pi, 4000)
r = A * (1+ np.cos(theta))
x = r * np.sin(theta)
y = r * np.cos(theta)

plt.axis('off');
plt.axis('equal');
plt.plot(x,y);
```

grafica segunda funci√≥n polar

```python
from re import A
from sympy import symbols
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(1, 4 * np.pi, 4000)
r = (A/theta)

plt.polar(theta, r, marker='+',color=(.2,.4,.5,.3) );
plt.show();
```

tarea sumatoria

```python
from re import A
from sympy import symbols, sin, pi
from sympy.plotting import plot
x = symbols('x')


fx = 1
for k in range(100):
  fx = fx + ((sin (2*pi*(k**2)*x))/(4))
n = 100  # N√∫mero de t√©rminos en la sumatoria
suma = 0

for k in range(1, n + 1):
  suma += (x**2)/(2*k)


print(fx)
print(suma)
```

tarea integral

```python
from sympy import symbols, Integral, sqrt

x = symbols('x')
intsen  = Integral(sqrt(1+x**3),(x,-1,1))
print("El resultado de la integral es:", intsen.evalf())
```
